```python
def find_count_mod(word, index, df):
    if df.loc[index]['ex_scroll'] == 0:
        return 0
    count = 0
    for i in range(0, index):
        if word in df.loc[i]['Name']:
            count += 1
    for j in range(index + 1, len(df)):
        if word in df.loc[j]['Name']:
            count += 1
    return count
```
